### Step 4 - Data Preprocessing

```python
df_con = df.select_dtypes(include=[np.number])
plt.figure(figsize=[12,8])
sns.boxplot(x='variable', y='value', data= pd.melt(df_con), width=1)
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df['Sales'])
plt.xlim(-100,5000)
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['Quantity'])
```

```python
plt.figure(figsize=(11,7))
sns.countplot(x=df['Ship Mode'])
```

```python
sns.pairplot(df, hue="Ship Mode")
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['Segment'])
```

```python
sns.pairplot(df, hue="Segment")
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['Category'])
```

```python
plt.figure(figsize=(8,6))
plt.bar('Category','Sales',data=df)
plt.title('Sales vs Category')
plt.xlabel('Category')
plt.ylabel('Sales')
```

```python
sns.countplot(x='Category',hue='Region',data=df)
plt.title("Region wise Category")
```

```python
plt.figure(figsize=[8,6])
sns.barplot(x="Category", y="Profit", data=df)
```

```python
plt.figure(figsize=(18,9))
sns.countplot(x=df['Sub-Category'])
plt.xticks()
```

```python
plt.figure(figsize=(18,8))
sns.countplot(x='Sub-Category',hue='Region',data=df)
plt.title("Region wise Sub-Category")
```

```python
print(df['State'].value_counts())
plt.figure(figsize=(18,9))
sns.countplot(x=df['State'])
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(19,8))
sns.barplot(x=df['Sub-Category'],y=df['Profit'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct= '%.2f%%')
```

```python
df.groupby(['Segment']).sum()[['Sales','Profit']].plot.bar()
plt.xticks(rotation=45)
plt.title("Segment Wise Sales Vs Profits")
```

```python
plt.figure(figsize= (18,9))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per sub-category')
plt.show()
```
